```python
import csv

def process_salary_data(csv_file):
    try:
        # Create a dictionary to store the salary data
        salary_data = {}
        
        # Open the CSV file and read it
        with open(csv_file, mode='r') as file:
            reader = csv.DictReader(file)  # Use DictReader to read rows as dictionaries
            
            # Create a function to safely convert strings to floats
            def safe_float(value):
                try:
                    return float(value)
                except ValueError:
                    return 0.0  # Default value for invalid or non-numeric inputs
            
            # Process each row
            for row in reader:
                # Use 'EmployeeName' as the reference and the rest of the data as the value
                employee_name = row['EmployeeName']
                salary_data[employee_name] = {
                    'BasePay': safe_float(row.get('BasePay', 0)),
                    'OvertimePay': safe_float(row.get('OvertimePay', 0)),
                    'OtherPay': safe_float(row.get('OtherPay', 0)),
                    'TotalPay': safe_float(row.get('BasePay', 0)) 
                                + safe_float(row.get('OvertimePay', 0)) 
                                + safe_float(row.get('OtherPay', 0)),
                }
                
        return salary_data
    except FileNotFoundError:
        return f"The file '{csv_file}' was not found."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage
csv_path = r'C:\Users\arubombora\PycharmProjects\Assign2\SF-SalaryData.csv'

result = process_salary_data(csv_path)

if isinstance(result, dict):
    # Example: Print all employee names and their total pay
    for employee, details in result.items():
        print(f"Employee: {employee}, TotalPay: {details['TotalPay']}")
else:
    # If an error occurred, print the error message
    print(result)
```

```python
import csv
import os
import zipfile

def save_to_csv_and_zip(salary_data, csv_filename, zip_filename):
    try:
        # Create the output directory if it doesn't exist
        os.makedirs(zip_filename, exist_ok=True)
        
        # Write the employee details to a CSV file
        with open(csv_filename, mode='w', newline='') as file:
            fieldnames = ['EmployeeName', 'BasePay', 'OvertimePay', 'OtherPay', 'TotalPay', 'Benefits']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            
            # Write the header row
            writer.writeheader()
            
            # Write the employee data
            for employee, details in salary_data.items():
                row = {
                    'EmployeeName': employee,
                    'BasePay': details['BasePay'],
                    'OvertimePay': details['OvertimePay'],
                    'OtherPay': details['OtherPay'],
                    'TotalPay': details['TotalPay'],
                    'Benefits': details.get('Benefits', 'Not Provided')
                }
                writer.writerow(row)

        # Zip the CSV file
        with zipfile.ZipFile(os.path.join(zip_filename, 'Employee Profile.zip'), 'w', zipfile.ZIP_DEFLATED) as zipf:
            zipf.write(csv_filename, os.path.basename(csv_filename))

        # Clean up: Delete the temporary CSV file after zipping
        os.remove(csv_filename)
        print(f"The CSV file has been saved and zipped as 'Employee Profile.zip' in the folder '{zip_filename}'")

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
csv_path = 'employee_details.csv'
zip_folder = r'C:\Users\arubombora\PycharmProjects\Assign2\Employee Profile'  
# Assuming `result` is the dictionary with employee details 
save_to_csv_and_zip(result, csv_path, zip_folder)

```
